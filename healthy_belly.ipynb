```
import random

# --- 1. KNOWLEDGE BASES ---

# Food Database: Density (g/cm3), Oxidative Load (AGEs/100g), Antioxidant Power (ORAC/100g)
FOOD_DB = {
    "Fried Chicken": {"density": 0.8, "ages_per_100g": 8000, "orac_per_100g": 0, "type": "heavy"},
    "French Fries": {"density": 0.3, "ages_per_100g": 1500, "orac_per_100g": 0, "type": "heavy"},
    "Blueberries": {"density": 0.6, "ages_per_100g": 0, "orac_per_100g": 4669, "type": "light"},
    "Papaya": {"density": 0.9, "ages_per_100g": 0, "orac_per_100g": 300, "type": "healing"}
}
```

```
# Asana Database for Bloating & Digestion
ASANA_DB = {
    "Vajrasana (Thunderbolt Pose)": "Sit on your heels for 5-10 minutes post-meal to increase blood flow to the stomach.",
    "Pawanmuktasana (Wind-Relieving Pose)": "Lie on your back and hug your knees to your chest to relieve trapped gas and bloating.",
    "Supta Baddha Konasana (Reclining Bound Angle)": "Relaxes the abdominal muscles and stimulates digestion."
}
```

```
# --- 2. THE VISION INTERFACE (Simulated) ---

def simulate_camera_vision(photo_path):
    """
    In reality, this uses a CNN and a reference object to gauge volume.
    Here, we simulate the model detecting a food and its estimated volume in cm^3.
    """
    print(f"📸 Scanning {photo_path}...")
    # Simulating detection of a large plate of fried chicken
    return {"food_name": "Fried Chicken", "volume_cm3": 400}
```

```
# --- 3. THE ANALYTICS ENGINE ---

def analyze_meal(vision_data):
    food_name = vision_data["food_name"]
    volume = vision_data["volume_cm3"]

    if food_name not in FOOD_DB:
        return "Food not recognized."

    stats = FOOD_DB[food_name]

    # Calculate portion weight: Volume (cm3) * Density (g/cm3)
    estimated_weight_g = volume * stats["density"]

    # Calculate total loads based on weight
    total_ages = (estimated_weight_g / 100) * stats["ages_per_100g"]
    total_orac = (estimated_weight_g / 100) * stats["orac_per_100g"]

    print("\n--- 🍽️ MEAL ANALYSIS ---")
    print(f"Detected: {food_name}")
    print(f"Estimated Portion: {estimated_weight_g:.0f} grams")
    print(f"Oxidative Load (AGEs): {total_ages:.0f} units")
    print(f"Antioxidant Power (ORAC): {total_orac:.0f} units")

    return generate_prescription(food_name, estimated_weight_g, total_ages, total_orac)
```

```
# --- 4. THE RECOMMENDATION ENGINE ---

def generate_prescription(food_name, weight, ages, orac):
    print("\n--- 🩺 YOUR CARE PRESCRIPTION ---")

    if ages > 5000:
        print("⚠️ HIGH OXIDATIVE STRESS & BLOAT RISK DETECTED.")
        print("This meal is heavy in complex fats and advanced glycation end products.")

        # Prescribe balancing food
        print("\n🥝 Dietary Prescription:")
        print("Consume a bowl of Papaya or Kiwi within the next hour. These contain strong digestive enzymes (Papain/Actinidin) to break down the lipids and counteract the oxidative stress.")

        # Prescribe Asana
        asana_name, asana_desc = random.choice(list(ASANA_DB.items()))
        print(f"\n🧘 Yoga Prescription for Bloating:")
        print(f"{asana_name}: {asana_desc}")

    elif orac > 2000:
        print("✅ Excellent choice! This meal is actively repairing cellular damage and reducing inflammation.")
    else:
        print("⚖️ A balanced meal, but consider adding a side of berries to boost your antioxidant intake.")


# --- RUN THE APP ---
if __name__ == "__main__":
    # User uploads a photo of their meal
    raw_image_data = "user_camera_upload_001.jpg"

    # Run pipeline
    detected_data = simulate_camera_vision(raw_image_data)
    analyze_meal(detected_data)
```
